```python
## using train and test method

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Loading the dataset
df = pd.read_csv(r'C:\Users\hariskumar uta\Downloads\Housing (1).csv')
df.head()

df.bedrooms.isnull().sum()

# represent the relations
sns.lmplot(x='area',y='price',data=df)
sns.lmplot(x='bedrooms',y='price',data=df)

#removing outliers
import numpy as np
q3 = df['price'].quantile(0.75)
q1 = df['price'].quantile(0.25)

IQR = q3 - q1
IQR

ul = q3 + (1.5 * IQR)
ll = q1 - (1.5 * IQR)

upper = np.where(df['price'] >= ul)
lower = np.where(df['price'] <= ll)

# removing the outliers
df.drop(upper[0],inplace = True)
df.drop(lower[0],inplace = True)
print(ul,ll)

x = df.iloc[:,1:3].values
y = df.iloc[:,0].values

df.shape

# taking 80 as train and 20 as test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 10)


# Importing Linear Regression
from sklearn.linear_model import LinearRegression
reg = LinearRegression()

# Training the data
reg.fit(x_train,y_train)


y_pred = reg.predict(x_test)
y_pred

# checking the accurancy
from sklearn.metrics import r2_score
R_squared = r2_score(y_test,y_pred)
print(R_squared)

# predicting price with new data
reg.predict([[780,3]])
reg.predict([[1500,3]])
```
