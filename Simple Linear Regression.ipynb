```python
# predicting the price of house in new york city.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load dataset
df = pd.read_csv(r'C:\Users\hariskumar uta\Downloads\homeprices (3).csv')

df.head()

# plot a scatter plot
plt.scatter(data=df,x='area',y='price')

from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(df[['area']],df['price'])

print(reg.predict([[3300]]))

print(reg.predict([[5000]]))

print(reg.coef_)

print(reg.intercept_)

y = reg.coef_ * 3300 + reg.intercept_
print(y)

from sklearn.metrics import r2_score
y_original = df.price
y_predicted = reg.predict(df[['area']])
R_square = r2_score(y_original,y_predicted)
print("r squared value ",R_square)

sns.lmplot(data=df,x='area',y='price')




```

```python
# Linear Regresson with test and train data
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r'C:\Users\hariskumar uta\Downloads\Salary_Data.csv')
df.head()

x = df.iloc[:,:-1].values
y = df.iloc[:,1].values

# scatter plot
# plt.scatter(x,y)

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3)

from sklearn.linear_model import LinearRegression
reg = LinearRegression()

reg.fit(x_train,y_train)

y_pred = reg.predict(x_test)
y_pred

from sklearn.metrics import r2_score
R_square = r2_score(y_test,y_pred)
print("R squared value ",R_square)

reg.predict([[11]])

# draw scatter plot and line plot
plt.scatter(x_train,y_train,color='red')
plt.plot(x_train,reg.predict(x_train),color='blue')
plt.title("Experience vs Salary")
plt.xlabel('Yrs of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
# canada per capita income
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r'C:\Users\hariskumar uta\Downloads\canada_per_capita_income (1).csv')
df.head()

x = df.iloc[:,:-1].values
y = df.iloc[:,1].values

# scatter plot
# plt.scatter(x,y)

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3)

from sklearn.linear_model import LinearRegression
reg = LinearRegression()

reg.fit(x_train,y_train)

y_pred = reg.predict(x_test)
y_pred

from sklearn.metrics import r2_score
R_square = r2_score(y_test,y_pred)
print("R squared value ",R_square)


# draw scatter plot and line plot
plt.scatter(x_train,y_train,color='red')
plt.plot(x_train,reg.predict(x_train),color='blue')
plt.title("PER CAPITA  INCOME OF CANADA")
plt.xlabel('Year')
plt.ylabel('Per capita income')
plt.show()
```
